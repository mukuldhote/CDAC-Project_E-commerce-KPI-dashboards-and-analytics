### Customer Lifetime Value (CLV) Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('clv_data.csv')
```

```python
df.head()
```

```python
import pandas as pd

# Calculate average purchase value
average_purchase_value = df['monetary_value'].mean()

# Calculate purchase frequency
purchase_frequency = df['frequency'].mean()

# Calculate average customer lifespan (in months)
average_customer_lifespan = df['customer_tenure'].mean()

# Calculate historical CLV
historical_clv = average_purchase_value * purchase_frequency * average_customer_lifespan

print('Historical CLV: ', round(historical_clv, 2),"brazillian real")
```

### Geospatial analysis: customer distribution

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'customer_data' with columns 'geolocation_lat', 'geolocation_lng', and other relevant columns
# Adjust column names based on your data

# Create a GeoDataFrame from the customer data
geometry = gpd.points_from_xy(df['geolocation_lng'], df['geolocation_lat'])
gdf = gpd.GeoDataFrame(df, geometry=geometry)

# Download a world map GeoDataFrame (you can use other geospatial data as needed)
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot customer distribution on the map
fig, ax = plt.subplots(figsize=(15, 10))
world.plot(ax=ax, color='lightgrey')
gdf.plot(ax=ax, marker='o', color='red', markersize=10, alpha=0.5)

# Set plot title and labels
plt.title('Customer Distribution', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)

# Show the plot
plt.show()

```

### Product Analysis:

```python
# Identify top-selling products
top_selling_products = df.groupby('product_id')['order_item_id'].count().sort_values(ascending=False).head(10)
```

```python
top_selling_products
```

```python

# Visualize top-selling products
plt.figure(figsize=(12, 6))
sns.barplot(x=top_selling_products.values, y=top_selling_products.index, palette='viridis')
plt.title('Top 10 Selling Products')
plt.xlabel('Number of Orders')
plt.ylabel('Product ID')
plt.show()

```

```python
# Analyze impact on customer behavior
product_behavior = df.groupby('product_id').agg({
    'order_id': 'nunique',
    'price': 'sum',
    'freight_value': 'sum',
    'product_weight_g': 'mean',
    'product_length_cm': 'mean',
    'product_height_cm': 'mean',
    'product_width_cm': 'mean',
    'review_score': 'mean',
}).sort_values(by='order_id', ascending=False).head(10)

product_corr = product_behavior.corr()

# Visualize impact on customer behavior
plt.figure(figsize=(12, 8))
sns.heatmap(product_behavior.corr(), annot=True, cmap='viridis')
plt.title('Correlation Heatmap - Product Impact on Customer Behavior')
plt.show()
```

```python
# Analyze product preferences of different customer segments
segment_product_preferences = df.groupby(['product_category_name_english', 'customer_state']).size().reset_index(name='counts')

# Filter out top N product categories
top_categories = segment_product_preferences.groupby('product_category_name_english')['counts'].sum().nlargest(10).index
filtered_data = segment_product_preferences[segment_product_preferences['product_category_name_english'].isin(top_categories)]

# Visualize product preferences among different customer segments
plt.figure(figsize=(20, 20))
sns.barplot(x='counts', y='product_category_name_english', hue='customer_state', data=filtered_data, palette='viridis')
plt.title('Top 10 Product Preferences Across Customer Segments')
plt.xlabel('Number of Orders')
plt.ylabel('Product Category')
plt.legend(title='Customer State')
plt.show()
```

#### Insights:
1. Review Score Correlations:
        Review_score has a negative correlation with late_deliveries (-0.231740). 
This indicates that higher instances of late deliveries might be associated with lower review scores. Timely delivery could be crucial for customer satisfaction.

2. Recency Correlations:
        recency has a very weak negative correlation with review_score (-0.068392). This suggests a slight negative association between recency (how recently a customer made a purchase) and the review score. However, the correlation is weak, so the relationship might not be significant.

3. Frequency and Monetary Value Correlations:
        frequency and monetary_value have a positive correlation of 0.286320. This indicates that customers who make more frequent purchases tend to have higher monetary values, which could be valuable for the business.
4. Customer Tenure and Other Features:
        customer_tenure has a positive correlation with both recency (0.025671) and frequency (0.118240). This suggests that longer-tenured customers might have made more recent purchases and tend to make purchases more frequently.

5. Number of Orders Correlations:
        num_orders has a strong positive correlation with both customer_tenure (0.604945) and frequency (0.257975). This makes sense, as customers who have been with the business for a longer time are likely to have placed more orders.

6. Late Deliveries Correlations:
        late_deliveries has a very weak negative correlation with customer_tenure (-0.016531). This suggests a slight negative association, but the correlation is weak, and customer tenure might not strongly influence the occurrence of late deliveries.

### Late_deliveries analysis

```python
late_del = df.copy()
```

```python
def state_encoding(state):
    if state in ['RS', 'SC', 'PR']:
        return 'southern'
    elif state in ['SP', 'RJ', 'MG', 'ES']:
        return 'southeastern'
    elif state in ['MT', 'MS', 'GO', 'DF']:
        return 'centralwestern'
    elif state in ['MA', 'PI', 'CE', 'RN', 'PB', 'PE', 'AL', 'SE', 'BA']:
        return 'northeastern'
    else: 
        return 'northern'
```

```python
late_del['customer_state'] = late_del['customer_state'].apply(state_encoding)
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='order_purchase_timestamp', y='late_deliveries', data=late_del, estimator='sum', ci=None, hue='customer_state', palette='viridis', markers=True, dashes=False)
plt.title('Trends in Late Deliveries Over Time by Customer State')
plt.xlabel('Order Purchase Timestamp')
plt.ylabel('Total Late Deliveries')
plt.xticks(rotation=45)
plt.legend(title='Customer State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Customer journey mapping:

```python
# Convert 'order_purchase_timestamp' to datetime
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])

# Resample data by quarter and sum the number of orders
orders_by_quarter = df.resample('Q', on='order_purchase_timestamp').sum()

# Map out the Typical Customer Journey - Quarterly
plt.figure(figsize=(14, 6))
sns.lineplot(x=orders_by_quarter.index, y='num_orders', data=orders_by_quarter, ci=None, color='blue')
plt.title('Customer Journey: Number of Orders Over Time (Quarterly)')
plt.xlabel('Quarter')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()

```
